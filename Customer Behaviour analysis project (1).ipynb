# 1.Customer Journey & Engagement Analysis

```python
#1.1 DROP-OFF points.

dropoff_data=""" Select c.CustomerID,c.CustomerName,P.productID,P.ProductName,cj.Stage
                 from customers as c
                 INNER JOIN customer_journey as cj ON c.CustomerID=cj.CustomerID
                 INNER JOIN products as p 
                 ON (p.ProductID=cj.ProductID)
                 WHERE cj.Customer_Action="Drop-off"
                 order by p.ProductID Desc;
"""                 
curr.execute(dropoff_data)
columns=curr.column_names
results=curr.fetchall()
data_frame=pd.DataFrame(results,columns=columns)
data_frame
```

```python
#1.2 Successful Conversion Data

conversion_data=""" select p.ProductID,P.ProductName
                     from project1.products as p
                     INNER JOIN (
                     select ProductID, COUNT(Customer_Action) AS Successful_Conversions
                     from project1.customer_journey
                     WHERE Customer_Action="Purchase"
                     Group by ProductID
                     ) AS cj
                     ON p.ProductID=cj.ProductID;
"""
curr.execute(conversion_data)
columns=curr.column_names
results=curr.fetchall()
data_frame=pd.DataFrame(results,columns=columns)
data_frame
```

```python
# 1.3 Average Duration per stage
avg_duration=""" select  CustomerID,ProductID, Stage,AVG(Duration) AS Average
                  from project1.customer_journey
                  Group by CustomerID,ProductID,Stage;
"""  
curr.execute(avg_duration)
columns=curr.column_names
results=curr.fetchall()
data_frame=pd.DataFrame(results,columns=columns)
data_frame
```

```python
# 2.1 Highest rated products

high_rating = """SELECT p.ProductID, p.ProductName, cr.Highest_Rating
    FROM project1.products AS p
    INNER JOIN (
        SELECT ProductID, MAX(Rating) AS Highest_Rating
        FROM project1.customer_reviews
        GROUP BY ProductID
        having max(rating)>3
    ) AS cr ON p.ProductID = cr.ProductID;
"""

curr.execute(high_rating)
columns=curr.column_names
results=curr.fetchall()
data_frame=pd.DataFrame(results,columns=columns)
data_frame
```

```python
#2.2 Lowest Rated products

low_rating=""" SELECT p.ProductID, p.ProductName, cr.Lowest_Rating
    FROM project1.products AS p
    INNER JOIN (
        SELECT ProductID, MIN(Rating) AS Lowest_Rating
        FROM project1.customer_reviews
        GROUP BY ProductID
        having min(rating)<4
    ) AS cr ON p.ProductID = cr.ProductID;
"""

curr.execute(low_rating)
columns=curr.column_names
results=curr.fetchall()
data_frame=pd.DataFrame(results,columns=columns)
data_frame
```

```python
# 2.3 sentiment analysis
# to view the customer review details for products

feedback=" select CustomerID, ReviewText from customer_reviews"
curr.execute(feedback)
columns=curr.column_names
feedback_data = curr.fetchall()
data_frame=pd.DataFrame(feedback_data,columns=columns)
data_frame
```

```python
#To Analyse the sentiment analysis of the customer

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()


def analyze_sentiment(feedback):
    sentiment_scores = analyzer.polarity_scores(feedback)
    compound_score = sentiment_scores['compound']
    
    # Determine sentiment as positive, negative, or neutral
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Process each feedback entry
for feedback_entry in feedback_data:
    customer_id = feedback_entry[0]
    feedback = str(feedback_entry[1])  
    sentiment = analyze_sentiment(feedback)
    print(f"Customer ID: {customer_id}, Sentiment: {sentiment}")
```

```python
# Add sentiment column to existing column

add_sentiment=""" ALTER TABLE customer_reviews ADD COLUMN Sentiment VARCHAR(10);
"""
curr.execute(add_sentiment)
db_connection.commit()
```

```python
# updating sentiment data 


for feedback_entry in feedback_data:
    customer_id = feedback_entry[0]
    feedback = feedback_entry[1]
    sentiment = analyze_sentiment(feedback)
    
    update_query = """
        UPDATE customer_reviews
        SET Sentiment = %s
        WHERE CustomerID = %s
    """
    curr.execute(update_query, (sentiment, customer_id))
    db_connection.commit()
```
